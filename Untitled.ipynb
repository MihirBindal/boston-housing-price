```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=420)
```

```python
rf = RandomForestRegressor()
```

```python
gsc = GridSearchCV(
        estimator=rf,
        param_grid={
            'n_estimators': [60,65,70,75,80],
            'max_features':[5,6,7,8],
            'max_depth': [5,6,7],
            'min_samples_leaf':[1,2]
        },
        cv=5, scoring='neg_mean_squared_error', verbose=0, n_jobs=-1)
```

```python
grid_result = gsc.fit(X_train, y_train)
best_params = grid_result.best_params_
best_params
```

```python
best_rf=RandomForestRegressor(max_depth=7, n_estimators=70, max_features=7, min_samples_leaf=1)
```

```python
cross_val_score(best_rf, X_train, y_train, cv=10, scoring='neg_median_absolute_error').mean()
```

```python
best_rf.fit(X_train,y_train)
y_predicted=best_rf.predict(X_train)
```

```python
mae = (np.sqrt(mean_absolute_error(y_train, y_predicted)))
mae
```

```python
y_predicted = best_rf.predict(X_test)
rmse = (np.sqrt(mean_absolute_error(y_test, y_predicted)))
rmse
```

```python
filename = 'finalized_model.pkl'
pickle.dump(best_rf, open(filename, 'wb'))
```

```python
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, y_test)
print(result)
```
